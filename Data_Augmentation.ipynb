```python

import soundfile as sf

## The implemented function 

def data_augmentation (path,path1):

  # Load the  orignial data
  waveform1, sample_rate1 = torchaudio.load(path)

  # Defining  effects

  effects = [
      ["speed", "0.6"],  # reduce the speed
      ["pitch",'4'],     #Pitch variation that meets the characteristics of Dysathric speech
      ['pad', '0', '1']]

  # Applying  effects

  waveform2, sample_rate2 = torchaudio.sox_effects.apply_effects_tensor(waveform1, sample_rate1, effects)

  path_save_1 = path1 + str(path.split('/')[-1])
  torchaudio.save( path_save_1, waveform2, sample_rate2) # Saved audio path

  ### Phase 2
  ## Adding delay

  waveform3, sample_rate3 = librosa.load(path_save_1)

  ###############    #############   ################

  duration = librosa.get_duration(filename= path_save_1) *1000
  
  # print(len(duration))

  sound = AudioSegment.from_wav(path_save_1)

  if duration < 4000:

    silence_second = sound[-1000:]
    first_second = sound[:1000]
    remaining_seconds = sound[1000:-1000]

    final_sound = first_second + silence_second + remaining_seconds

    y = final_sound.get_array_of_samples()
    sr = final_sound.frame_rate
    sf.write('/content/End/{file}'.format(file = str(path.split('/')[-1])), y, sr)  # Customize 

    print(str(path.split('/')[-1]))

    return final_sound
    
  else :

    silence_second = sound[-1000:]
    first_second = sound[:2000]
    remaining_seconds = sound[2000:-1000]

    final_sound = first_second + silence_second + remaining_seconds 

    y = final_sound.get_array_of_samples()
    sr = final_sound.frame_rate

    ## Final folder for saving the file

    sf.write('/content/drive/MyDrive/SoundProject/{file}'.format(file = str(path.split('/')[-1])), y, sr)  # Customize 

    return final_sound 
```
